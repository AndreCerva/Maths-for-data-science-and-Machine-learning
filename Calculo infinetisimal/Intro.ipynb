#### Funci√≥n

Relaci√≥n establecida entre dos conjuntos A y B que asigna a cada valor del conjunto A(variable independiente) un unico valor del conjunto B(variable dependiente)

#### Dominio de la funci√≥n

Todos los valores posibles de la variable independiente (valores q x va a tomar)

#### Rango de la funci√≥n

Conjunto de todos los valores posibles de la variable dependiente (valores q y va a tomar)

# Funci√≥n lineal

Tiene la forma de: ùëì(ùë•)=ùëöùë•+ùëè
 
donde  ùëö  y  ùëè   ‚ààùëÖ .

ùëö  puede ser calculada por:
ùëö=(ùë¶2‚àíùë¶1) / (ùë•2‚àíùë•1) y  ùëè, es el punto de corte con el eje  ùë¶.

Su dominio es  ùê∑ùëúùëöùëì=(‚àí‚àû,‚àû) . Su Rango es  ùêºùëöùëì=(‚àí‚àû,‚àû)

```python
#Puntos de ejemplo (2,4) y (3,5)
m=(5-4)/(3-2)
m
```

```python
#Rango en el que se va a evaluar la funcion
x=np.linspace(-10,10,1000) #Numero inicial, numero final, cantidad de espacios entre estos
x
```

```python
#Evaluamos la funci√≥n en el rango y con los datos
b=1 #Valor en el que nuestra funcipon cortara la linea de los ejes de la y, puede cambiar
y=lambda x: m*x+b
print(y(x))
```

```python
fig, ax = plt.subplots()
ax.plot(x,y(x),color='red')
ax.grid()
ax.axhline(y=0, color='black')
ax.axvline(x=0, color='black')
plt.show()
```

#### Composici√≥n de funciones

```python
x=np.linspace(-20,20,100)
cuadrado=lambda x: x**2#g
#print(cuadrado(cuadrado(x)))

seno=lambda x: np.sin(x)#f
#print(seno(x))

_,fig1=plt.subplots()
fig1.plot(x,cuadrado(x),color='red')
fig1.grid()
fig1.axhline(y=0, color='black')
fig1.axvline(x=0, color='black')
_,fig2=plt.subplots()
fig2.plot(x,seno(x),color='red')
fig2.grid()
fig2.axhline(y=0, color='black')
fig2.axvline(x=0, color='black')
plt.show()
```

```python
fog=lambda x: seno(cuadrado(x))
_,figfog=plt.subplots()
figfog.plot(x,fog(x),color='r')
figfog.grid()
figfog.axhline(y=0, color='black')
figfog.axvline(x=0, color='black')
plt.show()
```
