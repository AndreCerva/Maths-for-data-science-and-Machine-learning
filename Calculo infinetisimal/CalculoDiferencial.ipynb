#### Máximos y minimos

En una función los máximos son los valores más grandes y los minimos los valores más peueños que toma la función 
¿Cómo encontramos estos valores?
 1) Derivamos nuestra función
 2) Igualamos a 0
 3) Evaluamos en f(x) el valor optenido a igualar a 0
 4) Observamos si es un máximo o minimo con la segunda derivada, si es positiva es un minimo y si es negativa es un max
 De igual forma se puede evaluar la función en dos puntos cercanos x la izquierda y derecha a donde se tiene el punto mx o min y ver su comportamiento



```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Derivemos : 3x**2 + 5

1. Derivamos (Hallamos Primera derivada)
2. Igualamos a 0 y resolvemos
3. Reemplazamos en f(x)
4. Averiguamos si es un máximo o mínimo (Hallando la 2da derivada).

Primera derivada:

f(x)=3x^2+5
f'(x)=3(2x)+0
f'(x)=6x

Luego, igualo a 0 para hallar el punto en el eje de las x.

0=6x
x=0

Luego,
f(x)= 3(0)^2+5 => f(x)= (0)+5 =0+5 = 5

Punto: (0,5)
-----------------------

Segunda derivada:

f'(x)=6x
f''(x)=6(1)
f''(x)=+6
por lo tanto, tenemos un minimo

```python
x=np.linspace(-10,10,10000000)
y=lambda x: 3*x**2+5
fig,ax=plt.subplots()
ax.plot(x,y(x))
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('y=3x^2+5')
ax.axhline(0,color='black')
ax.axvline(0,color='black')
plt.show()
```

#### Actividad Única: Grafica f(x)=2x^2-4x-1, halla una escala donde se pueda apreciar que el valor del mínimo absoluto y obten la 2da derivada para verificar resultados.

```python
#f=2x**2-4x-1
y=lambda x: 2*x**2-4*x-1
fig,ax=plt.subplots()
ax.plot(x,y(x))
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('y=2x^2-4x-1')
ax.axhline(0,color='black')
ax.axvline(0,color='black')
plt.show()
```
